Trend of accidents per year/month

```python
# Ensure 'date' column is in datetime format
# This is necessary if the 'date' column was previously reverted to string type.
if not pd.api.types.is_datetime64_any_dtype(collision['date']):
    collision['date'] = pd.to_datetime(collision['date'], format='%d/%m/%Y', errors='coerce')

# Calculate accidents per month using chained operations
accidents_per_month = (
    collision['date'].dt.to_period('M')  # Convert to monthly period
    .value_counts()  # Count occurrences for each period
    .sort_index()  # Sort by period (chronologically)
    .reset_index(name='number_of_accidents')  # Convert to DataFrame
)
# The column created by reset_index is named 'date' (from the Series name), not 'index'.
# Rename the 'date' column to 'year_month'.
accidents_per_month.rename(columns={'date': 'year_month'}, inplace=True)
# Convert 'year_month' to string for proper plotting on x-axis
accidents_per_month['year_month'] = accidents_per_month['year_month'].astype(str)

plt.figure(figsize=(14, 7))
sns.lineplot(data=accidents_per_month, x='year_month', y='number_of_accidents', marker='o')
plt.title('Trend of Accidents per Year/Month')
plt.xlabel('Year-Month')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Heatmap of Accident Density by Latitude/Longitude

This heatmap visualizes the spatial distribution of accidents by using Kernel Density Estimation (KDE) based on the `latitude` and `longitude` coordinates from the `collision` dataset. Invalid coordinates (e.g., -1 or NaN) will be excluded for accurate representation.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out invalid latitude and longitude values (e.g., -1 or NaN)
# Assuming -1 indicates unknown values, and .dropna() handles NaNs
filtered_collision_coords = collision[(collision['latitude'] != -1) &
                                      (collision['longitude'] != -1) &
                                      (collision['latitude'].notna()) &
                                      (collision['longitude'].notna())]

plt.figure(figsize=(12, 10))
sns.kdeplot(x=filtered_collision_coords['longitude'],
            y=filtered_collision_coords['latitude'],
            fill=True, cmap='viridis',
            cbar=True, cbar_kws={'label': 'Density of Accidents'})
plt.title('Heatmap of Accident Density by Latitude/Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(False) # Turn off grid for better heatmap visualization
plt.tight_layout()
plt.show()
```

### Stacked Bar Chart of Vehicle Types vs. Accident Severity

This chart will display the distribution of vehicle types involved in accidents, broken down by the severity of the collision. To achieve this, we need to merge the `vehicle` and `collision` dataframes on a common key (`collision_index`) and then filter out unknown values in both `vehicle_type` and `collision_severity` before plotting.

```python
#Stacked bar chart of vehicle types vs. accident severity for the vehicle dataset
# Merge vehicle and collision dataframes on collision_index to get severity information
vehicle_collision = pd.merge(
    vehicle,
    collision[['collision_index', 'collision_severity']],
    on='collision_index',
    how='left'
)

# Filter out unknown/unspecified values (-1) from both columns
filtered_vehicle_collision = vehicle_collision[
    (vehicle_collision['vehicle_type'] != -1) &
    (vehicle_collision['collision_severity'] != -1)
]

# Create a cross-tabulation for counts
vehicle_severity_counts = pd.crosstab(
    filtered_vehicle_collision['vehicle_type'],
    filtered_vehicle_collision['collision_severity']
)

# Plotting the stacked bar chart
plt.figure(figsize=(15, 8))
vehicle_severity_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(15, 8))

plt.title('Vehicle Type Distribution by Collision Severity')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Vehicles Involved')

# Map severity codes to labels for better readability
severity_labels = {1: 'Fatal', 2: 'Serious', 3: 'Slight'}
plt.legend(title='Collision Severity', labels=[severity_labels[col] for col in vehicle_severity_counts.columns])

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Replace -1 in 'local_authority_district' with NaN in the collision DataFrame
collision['local_authority_district'] = collision['local_authority_district'].replace(-1, np.nan)

# Create the collision_details_dim
# Step 1: Select relevant columns from the original collision DataFrame, INCLUDING 'local_authority_district', and drop duplicates
collision_details_dim = collision[['local_authority_district', 'latitude', 'longitude', 'location_easting_osgr',
                                   'location_northing_osgr', 'collision_severity', 'weather_conditions']].drop_duplicates()

# Step 2: Add surrogate primary key
collision_details_dim['location_ID'] = collision_details_dim.reset_index(drop=True).index + 1

# Step 3: Reorder columns
collision_details_dim = collision_details_dim[['location_ID', 'local_authority_district', 'latitude', 'longitude',
                                               'location_easting_osgr', 'location_northing_osgr','weather_conditions']]
collision_details_dim.head()
```

```python
#Create a dimension for casualty
#Create the primary key casualty ID by combining casualty reference & collision index together
casualty['Casualty_ID'] = casualty['casualty_reference'].astype(str) + '-' + casualty['collision_index'].astype(str)

#Build the dimension table
casualty_dim = casualty[['Casualty_ID','age_of_casualty','sex_of_casualty','casualty_type','casualty_severity']].drop_duplicates()
casualty_dim.head()
```

```python
#Create a dimension for vehicle
#Create the primary key vehicle ID by combining vehicle reference & collision index together
vehicle['Vehicle_ID'] = vehicle['vehicle_reference'].astype(str) + '-' + vehicle['collision_index'].astype(str)

#Build the dimension table
vehicle_dim = vehicle[['Vehicle_ID','vehicle_type','age_of_vehicle','engine_capacity_cc','age_of_driver']].drop_duplicates()
vehicle_dim.head()
```

```python
#Create a dimension for time from the collision dataset
#Create the primary key Time_ID by combining date & collision index
collision['Time_ID'] = collision['date'].astype(str) + '-' + collision['collision_index'].astype(str)

collision['day_of_week'].unique()
day_of_week_map = {1:"Monday",
2 : "Tuesday",
3 : "Wednesday",
4 : "Thursday",
5 : "Friday",
6 : "Saturday",
7 : "Sunday"
}


#Build the dimension table
collision['day_of_week'] = collision['day_of_week'].map(day_of_week_map)
time_dim = collision[['Time_ID','date','day_of_week','time']].drop_duplicates()
time_dim.head()
```

The fact table is created centering collision. Collisions are the main reason why the casualties exist and also why the vehicles are checked as well

```python
# Define the columns that uniquely identify a 'location' in collision_details_dim.
location_identifying_cols = [
    'local_authority_district', 'latitude', 'longitude',
    'location_easting_osgr', 'location_northing_osgr'
]

# Create the fact table by chaining merge operations to bring in all necessary IDs and measures.
# Start with the collision data, which contains collision_index, Time_ID, and measures.
fact_table = (
    collision
    .merge(
        collision_details_dim[location_identifying_cols + ['location_ID']],
        on=location_identifying_cols,
        how='left'
    )
    .merge(
        casualty[['collision_index', 'casualty_reference', 'Casualty_ID']],
        on='collision_index', # Merge casualty information based on collision_index
        how='left'
    )
    .merge(
        vehicle[['collision_index', 'vehicle_reference', 'Vehicle_ID']],
        on='collision_index', # Merge vehicle information based on collision_index
        how='left'
    )
    # Select the final columns for the fact table
    [['collision_index', 'location_ID', 'Casualty_ID', 'Vehicle_ID', 'Time_ID',
      'number_of_vehicles', 'number_of_casualties', 'collision_severity']]
)

# Display the head of the fact table to verify
fact_table.head()
```
